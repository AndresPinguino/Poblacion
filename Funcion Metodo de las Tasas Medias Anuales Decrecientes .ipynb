```python
def metodo_tasas_medias_anuales_decrecientes(poblacion, censos, año_actual):

    tiempo = [0] * 7


    for i in range(len(censos)):
        tiempo[i] = censos[i]

    tiempo[3]=año_actual
    tiempo[4]=tiempo[3]+2
    tiempo[5]=tiempo[4]+10
    tiempo[6]=tiempo[5]+10


    datos = [0] * 7


    for i in range(len(poblacion)):
        datos[i] = poblacion[i]


    iI=((datos[1]/datos[0])**(1/(tiempo[1]-tiempo[0])))-1
    iII=((datos[2]/datos[1])**(1/(tiempo[2]-tiempo[1])))-1


    i1=iII
    i1=round(i1, 4)


    Pa=datos[2]*(1+i1)**(tiempo[3]-tiempo[2])
    Pa=round(Pa,0)
    Pa=int(Pa)


    P0=Pa*(1+i1)**(tiempo[4]-tiempo[3])
    P0=round(P0,0)
    P0=int(P0)


    P10=P0*(1+i1)**(tiempo[5]-tiempo[4])
    P10=round(P10,0)
    P10=int(P10)


    i2=(iI+iII)/2
    i2=round(i2,4)


    if (i2<=i1):
        i2=i2
    else:
        i2=i1

    P20=P10*(1+i2)**(tiempo[6]-tiempo[5])
    P20=round(P20,0)
    P20=int(P20)


    datos[3]=Pa
    datos[4]=P0
    datos[5]=P10
    datos[6]=P20

    #print (tiempo)
    #print (datos)

    return datos
```

```python
print (tiempo)
proyecciones = [0] * len(poblaciones)

for i in range(len(poblaciones)):
    proyecciones[i] = metodo_tasas_medias_anuales_decrecientes(poblaciones[i], censos, año_actual)
    print (proyecciones[i])
```
